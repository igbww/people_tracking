# **MOT**

**Параметры аннотаций MOT**

ID - это уникальный идентификатор изображения.

gtboxes - это массив аннотаций, которые содержат информацию о каждом объекте, найденном на изображении.

gtboxes": [{"fbox": [72, 202, 163, 503], "tag": "person", "hbox": [171, 208, 62, 83]

fbox - это 4-мерный (xmin, ymin, xmax, ymax) случайный ящик, окружающий связь.

tag - это словесным описанием является "person" (в MOT).

hbox - 4-мерный (xmin, ymin, xmax, ymax) bounding box, окружающим head box.

extra - дополнительная информация об объекте, такая как box_id и уровень окклюзии.

vbox - 4-мерный (xmin, ymin, xmax, ymax) bounding box, окружающим visible box.

head_attr - атрибуты головы, такие как флаг игнорирования и уровень окклюзии.

Окклюзия в MOTOR – это правило, которое используется для уменьшения размера базы данных. Оно указывает, что если условие A является истинным, то условие B недопустимо.

gtboxes - это аннотация, которая используется для описания рамок границы объектов, распознаваемых на изображении.

fbox - это рамка, которую следует использовать, чтобы определить рамку, включающую все 4 угловые точки (x1, y1, x2, y2).

tag - это связка, которую следует использов

Параметры аннотации fbox и hbox используются для указания размеров объектов на изображении. fbox указывает ширину, высоту, x-координату и y-координату, hbox - ширину, высоту, x-координату центра и y-координату центра.

```python
import os
from pathlib import Path

path = '/content/t/train/labels/'
list_0 = []
for file in os.listdir('/content/t/train/labels'):
  result = os.stat(path + file)
  if result.st_size < 35: # Если размер файла меньше 35 кб (37и больше кб весит файл с записью )
    list_0.append(file)  #.split('.')[0]

# Удаляем пустые файлы
for file in list_0:
    os.remove('/content/t/train/labels/'+file)
len(list_0)
```

**Функция для формирования gt.txt №2**

```python
def yolo_to_mot(yolo_data, file_name):
    # Разбиваем строку с данными YOLO на составляющие
    data = yolo_data.split(' ')
    if len(data) == 5:
        class_id, x, y, w, h = data
        x = int(float(x) * 1280)
        y = int(float(y) * 1280)
        w = int((float(w) * 1280)/2)
        h = int((float(h) * 1280)/2)
        
    # Формируем MOT-данные. ID указывать необязательно.  
    mot_data = int(file_name.split('.')[0]), int(-1), x, y, w, h, int(1), int(class_id), float(1.0)   #,-1 int(class_id)
    return mot_data
```

# **Конвертация анотаций из YOLO в FairMOT**

```python
import xml.etree.ElementTree as ET


def convert_yolo_to_fairmot(yolo_annotations_file_path, fairmot_annotations_file_path):
    # Открываем файл аннотаций YOLO
    with open(yolo_annotations_file_path, 'r') as f:
        yolo_annotations = f.readlines()
    
    # Создаем новый XML-файл формата FairMOT
    root = ET.Element('annotations')
    
    # Обходим все записи аннотаций YOLO
    for yolo_annotation in yolo_annotations:
        # Разбиваем строку аннотации на координаты, класс и центр тяжести
        x, y, width, height, class_id, center_x, center_y = map(float, yolo_annotation.split())
        
        # Создаем новую запись XML-файла формата FairMOT
        object_elem = ET.SubElement(root, 'object')
        
        # Указываем координаты объекта
        bbox_elem = ET.SubElement(object_elem, 'bndbox')
        ET.SubElement(bbox_elem, 'xmin').text = str(int(x))
        ET.SubElement(bbox_elem, 'ymin').text = str(int(y))
        ET.SubElement(bbox_elem, 'xmax').text = str(int(x + width))
        ET.SubElement(bbox_elem, 'ymax').text = str(int(y + height))
        
        # Указываем класс объекта
        ET.SubElement(object_elem, 'name').text = str(int(class_id))
        
        # Указываем центр тяжести объекта
        ET.SubElement(object_elem, 'occlusion').text = str(0)
        ET.SubElement(object_elem, 'generated').text = str(0)
        ET.SubElement(object_elem, 'hide').text = str(0)
        ET.SubElement(object_elem, 'mask').text = str(0)
        ET.SubElement(object_elem, 'lost').text
```
