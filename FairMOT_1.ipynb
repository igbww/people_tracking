# **   YOLO  FairMOT**

```python
import xml.etree.ElementTree as ET


def convert_yolo_to_fairmot(yolo_annotations_file_path, fairmot_annotations_file_path):
    #    YOLO
    with open(yolo_annotations_file_path, 'r') as f:
        yolo_annotations = f.readlines()
    
    #   XML-  FairMOT
    root = ET.Element('annotations')
    
    #     YOLO
    for yolo_annotation in yolo_annotations:
        #     ,    
        x, y, width, height, class_id, center_x, center_y = map(float, yolo_annotation.split())
        
        #    XML-  FairMOT
        object_elem = ET.SubElement(root, 'object')
        
        #   
        bbox_elem = ET.SubElement(object_elem, 'bndbox')
        ET.SubElement(bbox_elem, 'xmin').text = str(int(x))
        ET.SubElement(bbox_elem, 'ymin').text = str(int(y))
        ET.SubElement(bbox_elem, 'xmax').text = str(int(x + width))
        ET.SubElement(bbox_elem, 'ymax').text = str(int(y + height))
        
        #   
        ET.SubElement(object_elem, 'name').text = str(int(class_id))
        
        #    
        ET.SubElement(object_elem, 'occlusion').text = str(0)
        ET.SubElement(object_elem, 'generated').text = str(0)
        ET.SubElement(object_elem, 'hide').text = str(0)
        ET.SubElement(object_elem, 'mask').text = str(0)
        ET.SubElement(object_elem, 'lost').text
```

```python
path_YOLO = '/content/138.txt'
path_MOT = '/content/138_MOT.xml'
convert_yolo_to_fairmot(path_YOLO, path_MOT)
```

```python
import os
import pandas as pd
train_dir = '/content/FairMOT/dataset/images/train/'
val_dir = '/content/FairMOT/dataset/images/val/'

training_files = os.listdir('/content/train/Images')
val_files = os.listdir('/content/val/Images')

training_df = pd.DataFrame(training_files,columns=['images'])
training_df.images = training_df.images.map(lambda x: train_dir + x)
val_df = pd.DataFrame(val_files,columns=['images'])
val_df.images = val_df.images.map(lambda x: val_dir + x)
```

```python
# ,    ,     to src/data
train_file = 'src/data/crowdhuman.train'
test_file = 'src/data/crowdhuman.val'
with open(train_file, 'w') as f:
    f.write("\n".join(training_df.images.values))
with open(test_file, 'w') as f:
    f.write("\n".join(val_df.images.values))
```
